### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

```python
df = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv', sep='\t')

df.to_csv('chipotle.tsv', sep='\t', index=False, header=False)

data = pd.read_csv('chipotle.tsv', sep='\t')

data
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = df
chipo
```

### Step 4. How many products cost more than $10.00?

```python
chipo['item_price'] = chipo['item_price'].astype(str)
chipo['item_price'] = chipo['item_price'].str.replace('$', '', regex=False)
dollarizer = lambda x: float(x)
chipo['item_price'] = chipo['item_price'].apply(dollarizer)

chipo1 = chipo[chipo['item_price'] > 10]
chipo1.shape[0]
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python
chipo2 = chipo[['item_name', 'item_price']]
chipo2
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values(by='item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
highest_price = chipo['item_price'].max()
chipo[chipo['item_price'] == highest_price]['quantity']
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo3 = chipo[chipo['item_name'] == 'Veggie Salad Bowl']
chipo3.shape[0]
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
chipo4 = chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)]
chipo4.shape[0]
```
