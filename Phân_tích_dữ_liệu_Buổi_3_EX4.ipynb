### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/pandas/drinks.csv).

### Step 3. Assign it to a variable called drinks.

```python
df = pd.read_csv('https://github.com/thieu1995/csv-files/blob/main/data/pandas/drinks.csv?raw=true')

df
```

```python
drinks = df

drinks
```

### Step 4. Which continent drinks more beer on average?

```python
drinks.groupby('continent')['beer_servings'].mean()
```

### Step 5. For each continent print the statistics for wine consumption.

```python
drinks.groupby('continent')['wine_servings'].describe()
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
mean_per_continent = df.groupby('continent')[['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']].mean()

print(mean_per_continent)
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
median_per_continent = df.groupby('continent')[['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']].median()

print(median_per_continent)
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
spirit_consumption = df.groupby('continent')['spirit_servings'].agg(['mean', 'min', 'max'])

spirit_consumption
```
