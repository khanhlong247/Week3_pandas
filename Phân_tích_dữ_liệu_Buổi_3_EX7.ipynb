### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
df = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv', sep='\t')

df.to_csv('chipotle.tsv', sep='\t', index=False, header=False)

chipo = df

chipo
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
item_counts = chipo.groupby('item_name')['quantity'].sum()

top5_items = item_counts.sort_values(ascending=False).head(5)

plt.figure(figsize=(10, 6))
top5_items.plot(kind='bar', color='skyblue')
plt.title('Top 5 most bought items')
plt.xlabel('Item name')
plt.ylabel('Quantity sold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
chipo['item_price'] = chipo['item_price'].replace('[\$,]', '', regex=True).astype(float)

order_data = chipo.groupby('order_id').agg({
    'quantity': 'sum',
    'item_price': 'sum'
}).reset_index()

plt.figure(figsize=(10, 6))
plt.scatter(order_data['item_price'], order_data['quantity'], color='green', alpha=0.6)
plt.title('Number of items ordered per order price')
plt.xlabel('Order price')
plt.ylabel('Number of items ordered')
plt.grid(True)
plt.tight_layout()
plt.show()
```
